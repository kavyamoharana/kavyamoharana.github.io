````python
chart1 = alt.Chart.from_dict({
    "data": {"url": "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv"},
    "mark": {"type":"bar"},
#   "width":"600",
#     "height":"400",
    "width":600,
    "height":400,
    "encoding": {
        "x": {
            "aggregate": "count",
            "type": "quantitative",
            "title": "Number of Buildings"
        },
        "y": {
           "field": "Agency Name",
            "type": "ordinal",
            "title": "Agency Name"
        },
        "color": {"value": "#9467bd"}
      }
})
````

````python
chart1
````

````python
chart2 = alt.Chart.from_dict({
    "data": {
        "url": "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/bfro_reports_fall2022.csv"
      },
    "params": [{
        "name": "grid",
        "select": "interval",
        "bind": "scales"
      }],
    "mark": "circle",
    "width":500,
    "height":500,
    "encoding": {
        "x": {
          "field": "temperature_low",
          "type": "quantitative",
          "scale": {"domain": [-15, 100]}
        },
        "y": {
          "field": "dew_point",
          "type": "quantitative",
          "scale": {"domain": [-20, 90]}
        },
        "color": {"field": "humidity", "type": "quantitative"}
    }
})
````

````python
chart2
````

````python
chart3 = alt.Chart.from_dict({
    "data": {"url": "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/bfro_reports_fall2022.csv"},
    "params": [
        {
          "name": "bfro",
          "select": {"type": "point", "fields": ["season", "state"]},
          "bind": {
            "season": {"input": "select", "options": [None, "Fall", "Winter", "Spring", "Summer", "Unknown"]},
            "state": {"input": "select", "options": [None, "Washington", "Ohio", "Florida", "Illinois", "Oregon", "West Virginia", "Kentucky", "Alabama", "Michigan", "Georgia"]}
          }
        },
        {
        "name": "grid",
        "select": "interval",
        "bind": "scales"
        }
    ],
    "mark": "circle",
    "width":500,
    "height":500,
    "encoding": {
        "x": {"field": "temperature_high", "type": "quantitative", "scale": {"domain": [-5, 110]}},
        "y": {"field": "dew_point", "type": "quantitative", "scale": {"domain": [-20, 100]}},
        "color": {
          "condition": {"param": "bfro", "field": "season", "type": "nominal"},
          "value": "grey"
        },
        "size": {"condition": {"param": "bfro", "value": 100}, "value": 50}
    }
})
````
