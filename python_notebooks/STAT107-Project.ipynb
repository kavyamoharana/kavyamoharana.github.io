## Dataset: "Top Hits Spotify from 2000-2019"
This dataset from kaggle contains audio statistics of the top 2000 songs on the music streaming platform Spotify from 2000 to 2019. Its columns contain general information about the songs as well as other properties and scores which Spotify categorizes its tracks by.

```python
import pandas as pd
df_spotify = pd.read_csv("spotify_hits.csv")
df_spotify
```

```python
# seeing how many years of release are in the dataset:
df_spotify["year"].unique()
```

```python
# I only want to analyze the songs which fall under the timeperiod 2000-2019
df = df_spotify[ (df_spotify["year"] >= 2000) & (df_spotify["year"] <= 2019) ]
df
```

**How many top hits does each year have?**

```python
df["year"].value_counts()
```

**What kind of genres are present in the top hits from 2000-2019?**

```python
import matplotlib.pyplot as plt

#top 10 genres with most number of songs:
genre_num = df["genre"].value_counts().nlargest(10)
genre_names = genre_num.index

#plotting bar graph:
plt.figure(figsize=(22, 11))
plt.title("Top 10 Genres of Hit Songs 2000-2019", fontsize=20)
plt.xlabel("Genre", fontsize=18)
plt.ylabel("Number of Songs", fontsize=18)
plt.bar(genre_names, genre_num, width=0.5)
```

**Who are the top artists?**

```python
artists = df["artist"].value_counts()
artists
```

```python
artists_num = artists.nlargest(10)
artists_name = artists_num.index

#plotting bar graph:
plt.figure(figsize=(14, 6))
plt.title("Top 10 Artists with Hit Songs 2000-2019", fontsize=16)
plt.xlabel("Artist", fontsize=12)
plt.ylabel("Number of Songs", fontsize=12)
plt.bar(artists_name, artists_num, width=0.4, color='mediumpurple')
```

```python
df_exp = df[ df["explicit"] == True ]
explicit = len(df_exp) / len(df)
not_explicit = 1 - explicit
percents = [explicit, not_explicit]
label = ["Explicit", "Not Explicit"]
plt.pie(percents, labels=label)
```

**What are the most and least popular songs?**

```python
most_popular = df.nlargest(10, "popularity")
most_popular
```

```python
least_popular = df.nsmallest(10, "popularity")
least_popular
```

**Is their popularity connected to factors like tempo or danceability?**

```python
print(most_popular["danceability"].mean())
print(most_popular["tempo"].mean())
print(most_popular["acousticness"].mean())
```

```python
print(least_popular["danceability"].mean())
print(least_popular["tempo"].mean())
print(least_popular["acousticness"].mean())
```
